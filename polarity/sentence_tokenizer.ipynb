```python
import pandas as pd
import numpy as np

# Load database as a dataframe
database_filepath= os.path.join(PATH, "data.csv")
database = pd.read_csv(database_filepath, na_values='')
ReviewText_col = pd.Series(database.review_txt.astype('unicode').values)
```

## Preprocess

```python
import spacy
nlp = spacy.load('fr_core_news_sm')
```

```python
text = list(ReviewText_col)
documents = nlp.pipe(text)
for doc in documents:
    print(list(doc.sents)[:20])
```

### Get long segmented text

```python
t=0
for parsed_review in nlp.pipe(text):
    for sent in parsed_review.sents:
        if len(sent)> 130 :
            print(sent)
            print('\n')
            t += 1
```

```python
print("we have {0} sentences > 130 caracters (toute une ligne) in total {1} reviews".format(t, len(text)))
```

# Chunking


```python
# text = list(ReviewText_col[:3])
prev_i = 0
new_txt = ""
for parsed_review in nlp.pipe(text):
        for sent in parsed_review.sents:
            if len(sent)> 50 : 
                added_sent = []
                for i, t in enumerate(sent):
                    if i != 0:
                        if t.pos_ == 'PRON' \
                        and (sent[i-1].pos_ not in ['ADP', 'PRON']) \
                        and (sent[i-1].text.lower() not in ['ne', "n'", "n", 'pas']):
                            added_sent = added_sent + [sent[prev_i:i]]
                            prev_i = i
                new_txt += ". ".join(map(str, added_sent))  + ". "    
            else:
            	new_txt += sent.text
        new_txt += u'\n'
```

```python
import codecs
segmented_review_txt_filepath = os.path.join(PATH, "polarity", "chunked_review_txt.txt")
with codecs.open(segmented_review_txt_filepath, 'w', encoding='utf_8') as f:
    f.write(new_txt)
```
